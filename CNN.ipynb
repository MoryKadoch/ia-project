```python
import keras
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.utils import to_categorical

# Load and preprocess the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

batch_size = 32
num_classes = 10
epochs = 5

# Define the CNN model
input = Input(shape=(28, 28, 1))
x = Conv2D(32, kernel_size=(3, 3), activation='relu')(input)
x = Conv2D(64, (3, 3), activation='relu')(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Dropout(0.25)(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

model = Model(input, output)

# Compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Print model summary
model.summary()

# Train the model
model.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, y_test))

# Evaluate the model
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
import keras
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.utils import to_categorical
from keras.callbacks import LearningRateScheduler

# Charger et prétraiter le jeu de données MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

batch_size = 32
num_classes = 10
epochs = 10

# Définir le modèle CNN
input = Input(shape=(28, 28, 1))
x = Conv2D(32, kernel_size=(3, 3), activation='relu')(input)
x = Conv2D(64, (3, 3), activation='relu')(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Dropout(0.25)(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

model = Model(input, output)

# Compiler le modèle
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Afficher un résumé du modèle
model.summary()

# Fonction de planification du taux d'apprentissage
def lr_schedule(epoch):
    lr = 0.001
    if epoch > 5:
        lr *= 0.1
    return lr

# Rappel de planification du taux d'apprentissage
lr_scheduler = LearningRateScheduler(lr_schedule)

# Entraîner le modèle avec le rappel de planification du taux d'apprentissage
model.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, y_test),
          callbacks=[lr_scheduler])

# Évaluer le modèle
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential, Model
from keras.layers import Flatten, Dense, Dropout
from keras.applications import VGG16
from keras.preprocessing.image import img_to_array, array_to_img
from keras.utils import to_categorical
from keras.callbacks import LearningRateScheduler

# Charger et prétraiter le jeu de données MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Convertir les images MNIST en images RGB et redimensionner au fur et à mesure de l'entraînement
X_train_resized = []
for img in X_train:
    img_rgb = img_to_array(array_to_img(img).convert("RGB"))
    img_resized = img_to_array(array_to_img(img_rgb).resize((224, 224)))
    X_train_resized.append(img_resized)

X_train_resized = np.array(X_train_resized)

X_test_resized = []
for img in X_test:
    img_rgb = img_to_array(array_to_img(img).convert("RGB"))
    img_resized = img_to_array(array_to_img(img_rgb).resize((224, 224)))
    X_test_resized.append(img_resized)

X_test_resized = np.array(X_test_resized)

# Créer le modèle VGG16 pré-entraîné
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Ajouter des couches fully connected au-dessus du modèle pré-entraîné
model = Sequential()
model.add(Flatten(input_shape=base_model.output_shape[1:]))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

# Créer le modèle complet pour le fine-tuning
model_fine_tune = Model(inputs=base_model.input, outputs=model(base_model.output))

# Geler les couches du modèle VGG16 pré-entraîné
for layer in base_model.layers:
    layer.trainable = False

# Compiler le modèle
model_fine_tune.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Afficher un résumé du modèle
model_fine_tune.summary()

# Fonction de planification du taux d'apprentissage
def lr_schedule(epoch):
    lr = 0.001
    if epoch > 5:
        lr *= 0.1
    return lr

# Rappel de planification du taux d'apprentissage
lr_scheduler = LearningRateScheduler(lr_schedule)

# Entraîner le modèle avec le rappel de planification du taux d'apprentissage
model_fine_tune.fit(X_train_resized, y_train,
                    batch_size=32,
                    epochs=10,
                    verbose=1,
                    validation_data=(X_test_resized, y_test),
                    callbacks=[lr_scheduler])
```
